# 1. 0~100 까지 난수 생성

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# 난수 생성
def CreateRandomData(Range, n):
    randomdata = np.zeros(n)
    
    for i in range(n):
        x = int(np.random.randint(0,Range))
        randomdata[i] = x
    return randomdata

def getHistogram(RandData, Range, n):
    Histogram = np.zeros(Range)
    
    for i in range(n):
        Histogram[RandData[i]] +=1
        
    return np.array(Histogram)
```

```python
randData = CreateRandomData(100,100)
print randData

Histo = getHistogram(randData,100,100)
print Histo
```

# 2. 0~100까지 난수 1000개 생성 및 히스토그램

```python
Range = 100
N = 1000

randData = CreateRandomData(Range,N)
Histo = getHistogram(randData,Range,N)

x = np.arange(Range)
plt.bar(x, Histo)
plt.xticks(list(range(0,Range)))
plt.xlabel('number')
plt.ylabel('frequency')
plt.grid()
plt.show()
```
