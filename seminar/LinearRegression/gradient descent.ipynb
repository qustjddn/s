# Gradient descent

```python
def getF(x,_a,_b):
    return _a*x + _b

y_target = 1.0
x = 0.0
a = 0.0;
b = 0.0
alpha = 0.01


for i in range(1000):
    error = y_target - getF(x,a,b)
    print "Squared Error = ",0.5*error*error
    
    Ea = error * -x
    Eb = error * -1
    
    a = a - alpha * Ea
    b = b - alpha * Eb   
    
print "y_target =", y_target
print "y =", getF(x,a,b)
```

# Gradient descent(Multiple sample)

```python
num = 3
y_t = [1.0,1.5,2.0]
x = [0.0,0.5,1.0]
alpha = 0.01
a = 0.0
b = 0.0

for i in range(1000):
    sqr_error_sum = 0.0
    Ea_sum = 0.0
    Eb_sum = 0.0
    
    for j in range(num):
        error = y_t[j] - getF(x[j],a,b)
        Ea = error * -x[j]
        Eb = error * -1
        
        sqr_error_sum += error *error
        Ea_sum += Ea
        Eb_sum += Eb
    
    print "squared Error =",0.5*sqr_error_sum
    
    a = a - alpha * (Ea_sum / num)
    b = b - alpha * (Eb_sum / num)

print "y_target = ",y_t
print "y =",[getF(x[0],a,b),getF(x[1],a,b),getF(x[2],a,b)]
```
